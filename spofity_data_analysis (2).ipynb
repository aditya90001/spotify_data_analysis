## Top 5 Genre BY Popularity

## Top 5 Track Name by Most Average Duration_Miniute

## Average Duration per Popularity Level

```python
#duration vs popularity
ax=df.groupby(["popularity_level"],as_index=False,observed=True)["duration_miniute"].mean().sort_values(by="duration_miniute",ascending=False).head(5)
```

```python
plt.figure(figsize=(12,4))
sns.barplot(data=ax,x="popularity_level",y="duration_miniute",hue="duration_miniute",palette="coolwarm")
```

Keep songs in the 3.5–3.9 minute range for better chances of high popularity.

Avoid overly long tracks (>4 minutes), as they are less likely to reach "very high" popularity.
